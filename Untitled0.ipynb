```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("student_feedback.csv")
df.head()

```

```python
df['Overall Score'] = df.drop(columns=['Student ID']).mean(axis=1)
df[['Student ID', 'Overall Score']].head()
```

```python
sns.boxplot(y=df['Overall Score'])
plt.title("Overall Student Feedback Score")
plt.show()
```

```python
df.info()
df.columns

```

```python
df = df.drop(columns=['Unnamed: 0'])
df.isnull().sum()
```

```python
plt.figure()
sns.histplot(df.drop(columns=['Student ID']).mean(axis=1), bins=10)
plt.title("Overall Student Satisfaction Distribution")
plt.xlabel("Average Feedback Score")
plt.ylabel("Number of Students")
plt.show()
```

```python
avg_scores = df.drop(columns=['Student ID']).mean().sort_values(ascending=False)

avg_scores.plot(kind='barh')
plt.title("Average Score for Each Feedback Parameter")
plt.xlabel("Average Rating")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.drop(columns=['Student ID']).corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Between Feedback Parameters")
plt.show()
```

```python
plt.figure()
sns.countplot(x=df['Course recommendation based on relevance'])
plt.title("Course Recommendation Score Distribution")
plt.xlabel("Recommendation Score")
plt.ylabel("Number of Students")
plt.show()
```
